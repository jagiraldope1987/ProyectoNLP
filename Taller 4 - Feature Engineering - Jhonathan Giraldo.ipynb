```python
dialogos = pd.read_csv("C:/Users/jagir/OneDrive/Konrad/NPL/ProyectoNLP/Clase05092020/dialogos.csv")
dialogos

```

```python
dialogos['pre-procesado'] = dialogos.apply(lambda row: " ".join([word for word in re.sub(r"[\W\d_]+", " ", row.Locución.lower()).split() if word not in stopwords_sp]), axis = 1) 
dialogos
```

# Punto 2: Representación vectorial

- `[11 pts]` Crear una bolsa de palabras (BoW) del corpus usando la columna pre-procesada
- `[11 pts]` ¿Cuántas palabras hay en el vocabulario? (Usando la función de `sklearn`)

```python
vocabulario = dialogos["pre-procesado"].str.cat(sep=" ")##concatenar todos los textos
vocabulario = set(vocabulario.split())##quitar elementos repetidos en la lista con set
vocabulario = [(word,i+1) for i,word in enumerate(vocabulario)]##enumerar cada palabra de la lista comenzando desde 1
vocabulario
```

```python
# BoW usando la columna pre-procesada
def bow(frase):
    frase_bow = [0]*len(vocabulario)
    for word,i in vocabulario:
        if word in frase.split():
            frase_bow[i-1] = 1
    print(frase_bow)
```

```python
bow("Lo han entendido mal")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
bow_rep = count_vect.fit_transform(dialogos["pre-procesado"].values)

len(count_vect.vocabulary_)
```
