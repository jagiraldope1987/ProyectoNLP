```python
import numpy as np
 
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
 
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
 
import matplotlib.pyplot as plt
import PIL.Image

from IPython.display import display
```

```python
#Carga de la imagen
dragonball = PIL.Image.open("kdb.jpg")
display(dragonball)
```

```python
mask = np.array(dragonball)
mask
```

```python
#Cargar el texto
import pandas as pd
data = pd.read_csv('characters_list.csv')
data.head()
```

```python
palabras = data.description.str.cat(sep=" ")
wordcloud = WordCloud().generate(palabras)
 
plt.imshow(wordcloud)
plt.show()
```

```python
palabras = data.description.str.cat(sep=' ')
 
wordcloud = WordCloud(width=800, height=400,
                      max_font_size=150, max_words=250,
                      background_color='black', colormap='cool',
                      #stopwords=stopwords,
                      mask=mask, contour_width=0.5, contour_color='skyblue').generate(palabras) # https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
 
wordcloud.to_file("kdb.png")
plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
